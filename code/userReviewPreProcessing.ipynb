```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
import os
```

```python
os.listdir("../dataset")
```

```python
rawDataPath="../dataset/yummlyReviews"
print(len(os.listdir(rawDataPath)))
fileNames=[ele for ele in os.listdir(rawDataPath) if ele.endswith(".json") ]
len(fileNames)
```

```python
userReviewData=[]
for eachFile in fileNames:
    with open(rawDataPath+"/"+eachFile,"r") as f:
        reviews=json.load(f)
        userReviewData.extend(reviews)
```

```python
userReviewData
```

```python
#initializing dataframe
df=pd.DataFrame([userReviewData[i] for i in range(len(userReviewData))])
df
```

```python
df.to_csv("../dataset/yummlyUserReview.csv")
```
