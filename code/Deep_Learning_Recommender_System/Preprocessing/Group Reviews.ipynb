```python
data=pd.read_csv('/content/gdrive/My Drive/Final Year Project/DeepConn/Deep_Learning_Recommender_System/cleaned_reviews.csv').drop("Unnamed: 0", axis=1)
dat=data
```

```python
user_item_review = dat.drop("review", axis=1)
# user_item_review.head()
```

```python
def MakeList(x):
    T = tuple(x)
    if len(T) > 1:
        return list(T)
    else:
        return T[0]
```

```python
# DFGrouped = data.groupby(['user_id', 'food_id'])
# DF_Agg = DFGrouped.agg({'review' : MakeList})
```

```python
# # user_item_review = data.drop("review", axis=1)
# print(data.head())
# user_reviews = pd.pivot_table(data, index=["user_id", "food_id"], aggfunc=lambda x:x).drop("rating", axis=1)
# print(user_reviews.head())
# # food_reviews = pd.pivot_table(data,
# #                                index=["food_id", "user_id"],
# #                                aggfunc=lambda x: x).drop("rating", axis=1)
# d1=data.drop('food_id',axis=1)
# d2=data.drop('user_id',axis=1)
user_reviews = dat.groupby('user_id').agg({'review' : MakeList})
food_reviews = dat.groupby('food_id').agg({'review' : MakeList})
user_reviews
```

```python
def add_user_reviews(x):
    # ur = user_reviews.loc[x["user_id"]]
    ur = user_reviews.loc[x["user_id"]].values.tolist()
    # print(ur)
    fr = food_reviews.loc[x["food_id"]].values.tolist()
    x["userReviews"] = " ".join(list(map(lambda x: x[0], ur)))
    x["foodReviews"] = " ".join(list(map(lambda x: x[0], fr)))
    return x
```

```python
# s = pd.Timestamp(int(time()), unit="s")
grouped_cleaned_data = user_item_review.apply(add_user_reviews, axis=1)
# print(user_item_review)
# add_user_reviews(user_item_review)
# e = pd.Timestamp(int(time()), unit="s")
# print("took {}".format(e - s))
# user_item_review.head(10)
```
